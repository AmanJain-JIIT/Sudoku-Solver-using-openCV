```python
def check(nums):
            dict = {}
            for num in nums:
                if num in dict:
                    return False
                else:
                    if num<1 or num>9:
                        return False
                    else:
                        dict[num] = 1
            return True
        
def isValidSudoku(board):
        for row in board:
            nums = []
            for num in row:
                if num!=0:
                    nums.append(int(num))
            if check(nums)==False:
                return False
        for i in range(9):
            nums = []
            for j in range(9):
                num = board[j][i]
                if num!=0:
                    nums.append(int(num))
            if check(nums)==False:
                return False
        i = 0 
        j = 0
        while j<9 and i<9:
            nums = []
            for a in range(i, i+3):
                for b in range(j, j+3):
                    num = board[a][b]
                    if num!=0:
                        nums.append(int(num))
            if check(nums)==False:
                return False
            j+=3
            if j==9:
                j=0
                i+=3
            
        return True
```

```python
N=9
def isSafe(grid, row, col, num):
    for x in range(9):
        if grid[row][x] == num:
            return False
    for x in range(9):
        if grid[x][col] == num:
            return False
    startRow = row - row % 3
    startCol = col - col % 3
    for i in range(3):
        for j in range(3):
            if grid[i + startRow][j + startCol] == num:
                return False
    return True

def solveSudoku(grid, row, col):
    if (row == N - 1 and col == N):
        return True
    if col == N:
        row += 1
        col = 0
    if grid[row][col] > 0:
        return solveSudoku(grid, row, col + 1)
    for num in range(1, N + 1, 1):
        if isSafe(grid, row, col, num):
            grid[row][col] = num
            if solveSudoku(grid, row, col + 1):
                return True
        grid[row][col] = 0
    return False

def solve_puzzle(grid):
    if solveSudoku(grid, 0, 0):
        return grid
```
